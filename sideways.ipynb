```python
import RPi.GPIO as GPIO
import time

#def sideways():

# Set up GPIO pins
GPIO.setmode(GPIO.BOARD)

#input pins front
GPIO.setup(7, GPIO.OUT) 
GPIO.setup(11, GPIO.OUT) 
GPIO.setup(13, GPIO.OUT) 
GPIO.setup(15, GPIO.OUT) 

#input pins back
GPIO.setup(16, GPIO.OUT) 
GPIO.setup(18, GPIO.OUT) 
GPIO.setup(22, GPIO.OUT) 
GPIO.setup(24, GPIO.OUT) 

# Set motor speed
pwm1 = GPIO.PWM(7, 100)
pwm2 = GPIO.PWM(11, 100)
pwm3 = GPIO.PWM(13, 100)
pwm4 = GPIO.PWM(15, 100)

pwm5 = GPIO.PWM(16, 100)
pwm6 = GPIO.PWM(18, 100)
pwm7 = GPIO.PWM(22, 100)
pwm8 = GPIO.PWM(24, 100)

pwm1.start(0)
pwm2.start(0)
pwm3.start(0)
pwm4.start(0)

pwm5.start(0)
pwm6.start(0)
pwm7.start(0)
pwm8.start(0)

# Drive mecanum wheels forward
GPIO.output(7, True)
GPIO.output(11, False)
GPIO.output(13, False)
GPIO.output(15, True)

GPIO.output(16, False)
GPIO.output(18, True)
GPIO.output(22, True)
GPIO.output(24, False)

# Set motor speed for wheels
pwm1.ChangeDutyCycle(50) 
pwm2.ChangeDutyCycle(50) 
pwm3.ChangeDutyCycle(50) 
pwm4.ChangeDutyCycle(50) 

pwm5.ChangeDutyCycle(50) 
pwm6.ChangeDutyCycle(50) 
pwm7.ChangeDutyCycle(50) 
pwm8.ChangeDutyCycle(50) 

time.sleep(5) 

# Stop  wheels
GPIO.output(7, False)
GPIO.output(11, False)
GPIO.output(13, False)
GPIO.output(15, False)

GPIO.output(16, False)
GPIO.output(18, False)
GPIO.output(22, False)
GPIO.output(24, False)

pwm1.stop()
pwm2.stop()
pwm3.stop()
pwm4.stop()

pwm5.stop()
pwm6.stop()
pwm7.stop()
pwm8.stop()

GPIO.cleanup()
```
